<h2>Question 1 - Probabilities (10 Marks)</h2>

A box contains $n$ pairs of shoes ($2n$ shoes in total). If $2r$ (with the assumption that $2r \leq n$) shoes are selected at random, find the probability for the following scenarios:

<h3> Question 1a. (2 Marks)</h3>

$A_0$ = ‘No matching pair’ 

<h3> Answers 1.a</h3>

2r shoes are selected at random.

possible outcomes = choosing 2r shoes form 2n shoes

possible outcomes => $\binom{2n}{2r}$

To not have matching pairs, we can choose in $\binom{n}{2r}$ ways.

For each pair of shoes, there is a left and right shoe which can be chosen in $2^{2r}$ ways.

Thus our sample selection can be done in $\binom{n}{2r}$ * $2^{2r}$ ways.

p(‘No matching pair’) = $\frac{Sample selection}{Total possible outcomes}$

$P(A_{0})$= $\frac{\binom{n}{2r} * 2^{2r}}{\binom{2n}{2r}}$




<h3> Question 1b. (3 Marks)  </h3>

$A_1$ = ‘only one matching pair’

<h3> Answers 1.b</h3>

We have n pairs of shoes.

first select one matching pair from n in $\binom{n}{1}$ ways.

Then, we have left n-1 pairs from which we select 2r-2 distinct pairs of shoes.

=> $\binom{n-1}{2r-2}$ 

shoes can be either left or right after the first matching pair. we can do this in $2^{2r-2}$ ways.

possible outcomes => $\binom{2n}{2r}$

$P(A_{1})$ = $\frac{\binom{n}{1}*\binom{n-1}{2r-2}*2^{2r-2}}{\binom{2n}{2r}}$

<h3> Question 1c. (2 Marks)</h3>

$A_2$ = ‘exactly two matching pairs’

<h3> Answers 1.c</h3>

Exactly 2 are matching in $\binom{n}{2}$ ways. Rest are n-2 pairs from which we select 2r-4 shoes.

so this can be dome in $\binom{n-2}{2r-4}$ ways.

shoes can be left or right, which can be selected in $2^{2r-4}$ ways.

$P(A_{2})$ = $\frac{\binom{n}{2}*\binom{n-2}{2r-4}*2^{2r-4}}{\binom{2n}{2r}}$



<h3> Question 1d. (3 Marks)</h3>

$A_r$ = ‘exactly r matching pairs’

<h3> Answers 1.d </h3>

if exactly r matching pairs then we can do it in $\binom{n}{r}$ ways.

Rest are n-r pairs, and we are choosing 2r-2r = 0 shoes, so  $\binom{n-r}{0}$ ways = 1 way.

shoes are left or right, can be done in $2^{2r-2r}$ ways = 0 ways.

so, $P(A_{r})$ = $\frac{\binom{n}{r}}{\binom{2n}{2r}}$

The above table shows the joint probability distributioin of random variables R (from previous entropy calculation) and Y.

previously we get: 

$H($Chelsea Results$)$ = 1.446 bits of information = H(R)

from the joint probability distribution we can calculate the joint entropy of random variables R and Y.

H(R) = 1.446 bits

The exponential distribution is a probability distribution for non-negative real numbers. It is often used to model waiting or survival times. The version that we will look at has a probability density function of the form

$p(y | v) = exp(−e^{−v}y − v)$

where $y ∈ R_+$, i.e., y can take on the values of non-negative real numbers. In this form it has one parameter: a log-scale parameter $v$.  If a random variable follows a gamma distribution with log-scale $v$ we say that $Y ∼ Exp(v)$. If $Y ∼ Exp(v)$, then $E[Y] = e^v$ and $V[Y] = e^{2v}$.

<h3> Question 5.a (4 Marks) </h3>

Imagine we are given a sample of n observations $y = (y_1, . . . , y_n)$. Make sure to simplify your expression, and provide working.

<h3>Likelyhood of the given data </h3>

Likelyhood function = $\prod_{i = 1}^{n}p(y_{i} | v)$

$P(Y \mid v)$ = $\prod_{i = 1}^{n} exp(−e^{−v}y_{i} − v)$

=> $\prod_{i = 1}^{n} (exp(-v))(exp(-e^{-v}y_{i}))$


Likelyhood of the data = $(e^{-nv})(e^{-\sum_{i = 1}^{n}y_{i}e^{-v}})$

<h3> Question 5.b (2 Marks) </h3>

Take the negative logarithm of your likelihood expression and write down the negative loglikelihood of the data $y$ under the exponential model with log-scale $v$. Simplify this expression

<h3>Negative log likelyhood </h3>

Negative log likelyhood = $-\log((e^{-nv})(e^{-\sum_{i = 1}^{n}y_{i}e^{-v}}))$  

= $-(-nv - \sum_{i = 1}^{n}y_{i}(e^{-v}))$

= $nv + \sum_{i = 1}^{n}y_{i}(e^{-v})$


<h3> Question 5.c (4 Marks) </h3>

Derive the maximum likelihood estimator $\hat{v}$ for $v$. That is, find the value of $v$ that minimises the negative log-likelihood. You must provide working.

<h3> Deriving the maximum likelyhood estimator $\hat{v}$</h3>

$\hat{v}$ = ?

We have to find the value of $\hat{v}$ where the negative log likelyhood is minimised.

so, we take the derivative of $nv + \sum_{i = 1}^{n}y_{i}(e^{-v})$ and set it to 0 to find the maximum likelyhood estimator $\hat{v}$.


$\frac{d}{dv}(nv + \sum_{i = 1}^{n}y_{i}(e^{-v}))$ = 0

=> $n + (\sum_{i = 1}^{n}y_{i}(-e^{-v}))$ = 0

=> $n - (e^{-v})\sum_{i = 1}^{n}y_{i}$ = 0

=> $n = (e^{-v})\sum_{i = 1}^{n}y_{i}$

=> $e^{v} = \frac{\sum_{i = 1}^{n}y_{i}}{n}$

=> $\frac{\sum_{i = 1}^{n}y_{i}}{n} = \overline{y}$ 

=> $e^{v} = \overline{y}$ 

=> $\hat{v}$ = $\log(\overline{y})$  


maximum likelyhood estimator $\hat{v}$ = $\log(\overline{y})$

<h3> Approximate bias and variance of the maximum likelihood estimator $\hat{v}$ </h3>


Given n is sufficiently large, by following the asymptotic properties of maximum likelyhood estimators (namely the large sample property of MLE):

$\hat{v}$ is approximately $\hat{v} \sim {\sf N}(v, I^{-1}(v))$

Here, the inverse varianc I(v) is the fishers information.

I(v) = Expected value of second derivative of $nv + \sum_{i = 1}^{n}y_{i}(e^{-v})$ which is the negative log likelyhood function.

I(v) = $e^{-v}$E[($\sum_{i = 1}^{n}y_{i} $)] = $e^{-v}$*$n*e^{v}$ = n

$I^{-1}(v)$ = $\frac{1}{n}$ , this is the inverse of I(v)


$\hat{v}$ is approximately $\hat{v} \sim {\sf N}(v, \frac{1}{n})$

Approximate bias for $\hat{v}$ is E[$\hat{v}$ - v] $\approx$ 0
