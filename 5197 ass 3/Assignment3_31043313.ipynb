$\lambda .$ Thus, $Y=\sum_{i=1}^{n} X_{i}$ has a Poisson distribution with mean $n \lambda .$ Moreover, by the Central limit Theorem, $\bar{X}=Y / n$ has, approximately, a Normal $(\lambda, \lambda / n)$ distribution for large $n$. Show that for large values of $n,$ the distribution of
$$
2 \sqrt{n}\left(\sqrt{\frac{Y}{n}}-\sqrt{\lambda}\right)
$$
is independent of $\lambda$.

<h3> Answer 6 </h3>

Need to show that for large values of n, the distribution of $2 \sqrt{n}\left(\sqrt{\frac{Y}{n}}-\sqrt{\lambda}\right)$ is independent of $\lambda$.

for that it is required to show that for large values of n, the distribution of $2 \sqrt{n}\left(\sqrt{\frac{Y}{n}}-\sqrt{\lambda}\right)$ can be approximated by Normal(0,1) distribution.

from the weak law of large numbers,

$\frac{Y}{n}$ $\xrightarrow{\text{Pr}}$ $\lambda$ as n $\rightarrow$ ∞

hence, for large values of n, we can approximate $u(\frac{Y}{n})$ by its linear approximation around the value of $\lambda$.

$u(\frac{Y}{n})$ $\approx$ $u(\lambda)$ + $u^{'}(\lambda)(\frac{Y}{n} - \lambda)$

where, $u^{'}(\lambda)$ = $\frac{1}{2\sqrt\lambda}$

since, $u(\lambda) = \sqrt\lambda$, for the values of lambda greater than 0.

Therefore, for large values of n, we can see the following,

$\sqrt\frac{Y}{n}$ - $\sqrt\lambda$ $\approx$ $\frac{1}{2\sqrt n}$ * $(\frac{Y}{n} - \lambda)$ / $\sqrt\frac{\lambda}{n}$


next, using the CENTRAL LIMIT THEOREM:


$(\frac{Y}{n} - \lambda)$ / $\sqrt\frac{\lambda}{n}$ $\xrightarrow{\text{D}}$ $Z \sim {\sf Normal}(0,1)$ as n $\rightarrow ∞$

following from previous statements,

$2 \sqrt{n}\left(\sqrt{\frac{Y}{n}}-\sqrt{\lambda}\right)$ $\xrightarrow{\text{D}}$ $Z \sim {\sf Normal}(0,1)$ as n $\rightarrow ∞$


Thus, we can conclude that for large values of n, the distribution of $2 \sqrt{n}\left(\sqrt{\frac{Y}{n}}-\sqrt{\lambda}\right)$ tends to normalise to 1 as n tends to infinity.
```r
#q3
empty_model <- lm(Comb.FE ~ 1, fuel) # creating an empty model
summary(empty_model)

#using R step() function to perform a stepwise selection procedure with BIC penalty
#which is obtaited by taking the value of k = log(nrow(fuel))
#note: even though in the output it says AIC, it is BIC for taking a differnt k value in the function
step(empty_model, direction = "both", scope=formula(multi_fit), k = log(nrow(fuel))) # log(n) for BIC

#after bic pruning

BICpruned <- lm(Comb.FE ~ Eng.Displacement + Drive.Sys + No.Gears + 
                  Lockup.Torque.Converter + Aspiration, data = fuel)


summary(BICpruned)
anova(BICpruned)
```

<h4> Question 2.b (3 Marks) </h4>

Please write down rejection sampling process for this distribution here, let's say that the range is $[-2, 2]$ for the proposal distribution. (if you think using this answer for the previous part is a good idea, it is not)

<h4> Answer 2.b  </h4>

our f(X) = $e^{-x^{2}\pi}$

g(x) = $e^{-x}$

rejection sampling process:

 - Simulating $U \sim {\sf Uniform}(0, 1)$
 - Simulating a candidate X ~ g from the candidate density if U $\le$ $\frac{f(x)}{c*g(x)}$ then accept the candidate otherwise reject the candidate and go back to the beginning.
 
 
 
 




```r
#Rejection sampling code
set.seed(1)
f <- function(x){
  exp(-x^2*pi)
}

g <- function(x){exp(-x)}

range <- seq(-2,2, by=0.001)
 
range1 <- seq(0,2, by=0.001)
range2 <- seq(-2,0, by=0.001)


#plotting
plot(range,f(range), type = 'l', ylim=c(0,1.5))
lines(range1,g(range1))
lines(-range1,g(range1))


# min of c = max f/g

c = max(f(range)/g(range)) #constant to muliply g(x)
c



#plotting with constant

plot(range,f(range), type = 'l', ylim=c(0,1.5))
lines(range1,c * g(range1))
lines(-range1,c * g(range1))


#creating empty vector of upto 10000 spots for updating accepted values into, 0 values
accepted <- rep(0,10000)

#setting up the count for total attempts untill 10000 accepted attempts
sum=0
for(i in 1:length(accepted)){
  
  
  repeat{
    
    x<- -log(runif(1)) #inverse cdf of g(x)
    u<- runif(1)
    if(u <= f(x)/(c*g(x))) {
      
      break
      
      
    }
    sum=sum+1
    
  }
  if(runif(1)<0.5){
    Z=x
  }
  else{Z=-x}
  accepted[i]<- Z
  
  
}

hist(accepted,freq=F, breaks=50)
```

<h4> Question 2.c (4 Marks) </h4>

What would be the acceptance and rejection percentage here?

<h4> Answer 2.c  </h4>

in Rejection sampling:

P(Accepted X) = $\mathrm{P}(U \le \frac{f(x)}{c*g(x)})$

P(Accepted X) = $\int P(U \le \frac{f(x)}{c*g(x)} | X=x)dx$

P(Accepted X) = $\int \frac{f(x)}{c*g(x)}*g(x)dx$

P(Accepted X) = $\frac{1}{c}$ (note:1 divided by constant c gives us accepted percentage)

P(Accepted X) = $\frac{1}{1.08282936308266}$ = 0.923 = 92.3%

acceptence percentage = 92.3%

rejection percentage = 100 - 92.3 = 7.7%
